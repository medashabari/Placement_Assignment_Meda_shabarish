### 1.Get all Pokemons whose spawn rate is less than 5%

```python
df[df['spawn_chance'] < 0.05] # Creating a boolean mask inside with condition and accessing True Rows
```

### 2.Get all Pokemons that have less than 4 weaknesses

```python
df[df['weaknesses'].apply(lambda x : len(eval(x))) < 4] # applying the condition where pokemans have less than 4 weakness
```

### 3.Get all Pokemons that have no multipliers at all

```python
df[df['multipliers'].isnull()] # applying isnull condition to get pokemons that have no multpliers
```

### 5. Get all Pokemons whose spawn time is less than 300 seconds

```python
row=[]
for i in range(df.shape[0]):
    min_sec = df['spawn_time'].fillna('0').str.split(':')[i]
    if len(min_sec)==1 and int(min_sec[0])<300:
        row.append(i)
    elif (int(min_sec[0])*60+int(min_sec[1])) < 300:
        row.append(i)
df.iloc[row]
```
