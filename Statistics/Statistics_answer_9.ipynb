Q-9. Given the score of students in multiple exams

Test the hypothesis that the mean scores of all the students are the same. If not, name the
student with the highest score.

```python
import numpy as np
from scipy.stats import f_oneway
```

```python
# Given data
data = {
    'karan': [85, 90, 92],
    'Deepa': [70, 80, 85],
    'karthik': [90, 85, 88],
    'chandan': [75, 70, 75],
    'jeevan': [95, 92, 96]
}
```

```python
scores = np.array(list(data.values()))
```

```python
# Perform one-way ANOVA test
statistic, p_value = f_oneway(*scores)
```

```python
# Print the ANOVA test results
print("One-way ANOVA Test Results:")
print(f"Test Statistic: {statistic}")
print(f"P-value: {p_value}")
```

```python
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There are significant differences in the mean scores of the students.")
    # Find the student with the highest score
    highest_score_student = max(data, key=lambda x: sum(data[x]))
    print(f"The student with the highest score is: {highest_score_student}")
else:
    print("The mean scores of the students are not significantly different.")
```
